```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```
or_data['input1']=[1,1,0,0]
or_data['input2']=[1,0,1,0]
or_data['ouput']=[1,1,1,0]
```

```
and_data['input1']=[1,1,0,0]
and_data['input2']=[1,0,1,0]
and_data['ouput']=[1,0,0,0]
```

```
xor_data['input1']=[1,1,0,0]
xor_data['input2']=[1,0,1,0]
xor_data['ouput']=[0,1,1,0]
```

```
# Pass column names as strings to x, y, and hue, and the DataFrame to the data parameter
sns.scatterplot(x='input1', y='input2', hue='ouput', data=and_data, s=200)

or_data
# Pass column names as strings to x, y, and hue, and the DataFrame to the data parameter
sns.scatterplot(x='input1', y='input2', hue='ouput', data=or_data, s=200)

xor_data
# Pass column names as strings to x, y, and hue, and the DataFrame to the data parameter
sns.scatterplot(x='input1', y='input2', hue='ouput', data=xor_data, s=200)

from sklearn.linear_model import Perceptron
clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()

clf1.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
clf2.fit(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values)
clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)

clf1.coef_

clf1.intercept_

x=np.linspace(-1,1,5)
y=-x+1

plt.plot(x,y)
# Pass column names as strings to x, y, and hue, and the DataFrame to the data parameter
sns.scatterplot(x='input1', y='input2', hue='ouput', data=and_data, s=200)

clf2.coef_

clf2.intercept_

x1=np.linspace(-1,1,5)
y1=-x+0.5

plt.plot(x1,y1)
# Pass column names as strings to x, y, and hue, and the DataFrame to the data parameter
sns.scatterplot(x='input1', y='input2', hue='ouput', data=or_data, s=200)

clf3.coef_

clf3.intercept_

# Assuming plot_decision_regions is from mlxtend.plotting
# You might need to install it: !pip install mlxtend
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values, clf=clf3, legend=2)
```

Here are the key **limitations of a perceptron**:

1.  **Only handles linearly separable data**  can't solve problems like XOR.
2.  **Fixed output**  only binary (0 or 1), no probabilities or confidence.
3.  **No loss function**  uses a rule-based update, limiting learning flexibility.
4.  **No hidden layers**  can't model complex or non-linear patterns.
5.  **Converges only if data is separable**  otherwise, it may loop forever.
6.  **Not suitable for regression tasks**  can't predict continuous values.
